# Dataset Intentions
The ecommerce dataset does not contain demographic information, such as family income, personal income, age, state of residence, and race. The intention is to update the randomized 2000 Census information to 2023 inflation and then randomly assign the individuals in the Census dataset to the customers in the ecommerce dataset while maintaining gender and marital status.

```python
# Fix typos and rename columns
ecommerce_df = ecommerce_df.rename(columns={
                                  "HourSpendOnApp":"AppHours",
                                  "NumberOfDeviceRegistered":"RegisteredDevices",
                                  "PreferedOrderCat":"PreferredCategory",
                                  "NumberOfAddress":"NumberAddresses",
                                  "Complain":"Complaint",
                                  "OrderAmountHikeFromlastYear":"OrderIncreasePercent",
                                  "CouponUsed":"CouponsUsed",
                                  "DaySinceLastOrder":"DaysSinceOrder",
                                  "CashbackAmount":"AverageCashback"
})
ecommerce_df
```

# **Column Dictionary**

**Gender**: Gender

```python
# Find unique values of Marital Status to make sure both dataframes match
print(ecommerce_df['MaritalStatus'].unique())
print(census_df['marital_status'].unique())
```

The census data will be simplified so the following are reflected:


**Single**: never married / single , widowed

**Divorced**: divorced , separated

**Married**: married / spouse present , married / spouse absent

```python
# Replace all married instances
census_df['marital_status'].replace('Married/spouse present','Married',inplace=True)
census_df['marital_status'].replace('Married/spouse absent','Married',inplace=True)
# Replace all single instances
census_df['marital_status'].replace('Never married/single','Single',inplace=True)
census_df['marital_status'].replace('Widowed','Single',inplace=True)
# Replace all divorced instances
census_df['marital_status'].replace('Separated','Divorced',inplace=True)
# Confirm all instances replaced
print(census_df['marital_status'].unique())
```

```python
# Find unique values of Gender to make sure both dataframes match
print(ecommerce_df['Gender'].unique())
print(census_df['sex'].unique())
```

```python
# Inflate Census Income Amounts
for i in census_df.index:
  census_df.at[i,'2023_family_income'] = round(cpi.inflate((census_df.at[i,"total_family_income"]),(census_df.at[i,"census_year"])),2)
  census_df.at[i,'2023_personal_income'] = round(cpi.inflate((census_df.at[i,"total_personal_income"]),(census_df.at[i,"census_year"])),2)
census_df
```

```python
# Drop year and old values
census_df.drop(columns=['census_year','total_family_income','total_personal_income'],inplace=True)
census_df
```

```python
# Since customers must be 18+, remove all rows where individuals are under 18
census_df = census_df.loc[census_df["age"]>=18]
census_df
```

```python
# Some of the rows have no family income but do have personal income. This doesn't make sense, dropping these rows
census_df = census_df.loc[census_df["2023_personal_income"]<=census_df["2023_family_income"]]
census_df
```

```python
# Create combined columns for merging
census_df['merge'] = census_df[['sex','marital_status']].apply('-'.join,axis=1)
ecommerce_df['merge'] = ecommerce_df[['Gender','MaritalStatus']].apply('-'.join,axis=1)
```

```python
# Get unique values for census and ecommerce data
print(census_df['merge'].unique())
print(ecommerce_df['merge'].unique())
print(census_df['merge'].value_counts())
```

```python
# Partition census data out into unique merge column values
male_married_df = census_df.loc[census_df['merge']=='Male-Married',:]
male_single_df = census_df.loc[census_df['merge']=='Male-Single',:]
male_divorced_df = census_df.loc[census_df['merge']=='Male-Divorced',:]
female_married_df = census_df.loc[census_df['merge']=='Female-Married',:]
female_single_df = census_df.loc[census_df['merge']=='Female-Single',:]
female_divorced_df = census_df.loc[census_df['merge']=='Female-Divorced',:]

# Reset all indices
male_married_df.reset_index(inplace=True)
male_single_df.reset_index(inplace=True)
male_divorced_df.reset_index(inplace=True)
female_married_df.reset_index(inplace=True)
female_single_df.reset_index(inplace=True)
female_divorced_df.reset_index(inplace=True)

# Rename index columns
male_married_df.rename(columns={"index":"uniqueID"},inplace=True)
male_single_df.rename(columns={"index":"uniqueID"},inplace=True)
male_divorced_df.rename(columns={"index":"uniqueID"},inplace=True)
female_married_df.rename(columns={"index":"uniqueID"},inplace=True)
female_single_df.rename(columns={"index":"uniqueID"},inplace=True)
female_divorced_df.rename(columns={"index":"uniqueID"},inplace=True)
```

```python
# Randomly select individuals from the census data based on their marital status and gender
for i in ecommerce_df.index:
  if ecommerce_df.at[i,'merge'] == 'Male-Married':
    choice = np.random.randint(len(male_married_df))
    ecommerce_df.at[i,'uniqueID'] = male_married_df.at[choice,'uniqueID']
  elif ecommerce_df.at[i,'merge'] == 'Male-Single':
    choice = np.random.randint(len(male_single_df))
    ecommerce_df.at[i,'uniqueID'] = male_single_df.at[choice,'uniqueID']

  elif ecommerce_df.at[i,'merge'] == 'Male-Divorced':
    choice = np.random.randint(len(male_divorced_df))
    ecommerce_df.at[i,'uniqueID'] = male_divorced_df.at[choice,'uniqueID']

  elif ecommerce_df.at[i,'merge'] == 'Female-Married':
    choice = np.random.randint(len(female_married_df))
    ecommerce_df.at[i,'uniqueID'] = female_married_df.at[choice,'uniqueID']

  elif ecommerce_df.at[i,'merge'] == 'Female-Single':
    choice = np.random.randint(len(female_single_df))
    ecommerce_df.at[i,'uniqueID'] = female_single_df.at[choice,'uniqueID']

  elif ecommerce_df.at[i,'merge'] == 'Female-Divorced':
    choice = np.random.randint(len(female_divorced_df))
    ecommerce_df.at[i,'uniqueID'] = female_divorced_df.at[choice,'uniqueID']
```

```python
# Reset index and rename old index column to match other data
census_df.reset_index(inplace=True)
census_df.rename(columns={'index':'uniqueID'})
```

```python
# Merge the two datasets together based on the uniqueID from the census data. Remove all duplicative or extraneous columns and rename columns to standardized format
merged_df = pd.merge(ecommerce_df,census_df,left_on='uniqueID',right_on='index',how='left')
merged_df.drop(columns=['index','merge_x','merge_y','uniqueID','sex','marital_status'],inplace=True)
merged_df.rename(columns={'state_fips_code':'State',
                          'age':'Age',
                          'race_general':'Race',
                          '2023_family_income':'HouseholdIncome',
                          '2023_personal_income':'PersonalIncome'},inplace=True)
merged_df
```

```python
# Look at non-null values in dataset
merged_df.info()
```

```python
# Remove null values and save as new dataframe. Confirm no additional null values exist
cleaned_df = merged_df.dropna(how='any')
cleaned_df.info()
```

```python
# Write cleaned DF to new CSV file
cleaned_df.to_csv('cleaned_data.csv')
```
