```python
# Import Dependencies
import pandas as pd
```

```python
# Read CSV into DataFrame
df = pd.read_csv('/content/Customer-Churn-Model/Results/final_model_data.csv',index_col=0)
df
```

```python
# aggregate averages
df_agg = df.groupby('Model', as_index=False).agg(
    avg_precision = ('precision', 'mean'),
    avg_recall = ('recall', 'mean'),
    avg_f1_score = ('f1-score', 'mean'),
)

# back to orginal df
agg_df = pd.DataFrame(df_agg)
agg_df = agg_df.sort_values(by=['avg_f1_score'], ascending=False)
agg_df
```

```python
# SVM accuracy details for No Random Oversampling
df.drop(['Neighbors', 'support'], axis=1)

svm_over_false_df = df.loc[(df['Model'] == 'Support Vector Machines (SVM)') & (df['Oversample'] == False)]
svm_over_false_df = svm_over_false_df.sort_values(by=['Oversample', 'Scaler'])

# aggregate averages
svm_over_false_df  = svm_over_false_df .groupby('Scaler', as_index=False).agg(
                    avg_precision = ('precision', 'mean'),
                    avg_recall = ('recall', 'mean'),
                    avg_f1_score = ('f1-score', 'mean'),
)
svm_over_false_df = svm_over_false_df.sort_values(by='avg_f1_score', ascending=False)
svm_over_false_df
```

```python
# SVM accuracy details includes Random Oversampling

svm_over_true_df = df.loc[(df['Model'] == 'Support Vector Machines (SVM)') & (df['Oversample'] == True)]
svm_over_true_df = svm_over_true_df.sort_values(by=['Oversample', 'Scaler'])

# aggregate averages
svm_over_true_df  = svm_over_true_df .groupby('Scaler', as_index=False).agg(
                    avg_precision = ('precision', 'mean'),
                    avg_recall = ('recall', 'mean'),
                    avg_f1_score = ('f1-score', 'mean'),
)
svm_over_true_df = svm_over_true_df.sort_values(by='avg_f1_score', ascending=False)
svm_over_true_df
```
